```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Sea Surface Temperature (SST) of the years 1985, 2000 & 2020 in Corales del Rosario Park/Cartagena-Colombia.  

```python
dfa = pd.read_csv("sst_85.csv")
dfb = pd.read_csv("sst_00.csv")
dfc = pd.read_csv("sst_20.csv")
```

```python
dfa = dfa.drop("Unnamed: 0", axis=1)
dfb = dfb.drop("Unnamed: 0", axis=1)
dfc = dfc.drop("Unnamed: 0", axis=1)
```

# Year 1985 

```python
dfa.head()
```

```python
dfa.describe()
```

```python
plt.figure(figsize=(8,4))
plt.title('Temperature 1985', fontdict={'fontweight':'bold', 'fontsize': 18} )


plt.plot(dfa.time, dfa.sst1, 'r.-', label = 'sst 1,2,3,4,7,10')
plt.plot(dfa.time, dfa.sst11, 'y.-', label = 'sst 5,6,8,9,11')

plt.ylabel('Temperature (ºC)', {'fontsize': 14})
plt.xticks(dfa.time[::2])
plt.legend()

plt.show()
```

```python
# Correlations
plt.figure(figsize=(8,8))
sns.heatmap(dfa.corr(), annot=True, cmap='coolwarm')
```

# 1985, 2000 & 2020 

```python
# making a df with two variables (with the max temperature of 1985 and 2020)
dft = pd.concat([dfa.time, dfa.sst1, dfc.sst5 ], axis=1)
```

```python
dft
```

```python
plt.figure(figsize=(8,4))
plt.title('SST with max temperature in 1985 & 2020', fontdict={'fontweight':'bold', 'fontsize': 18} )


plt.plot(dft.time, dft.sst1, 'g.-', label= '1985')
plt.plot(dft.time, dft.sst5, 'r.-', label= '2020')

plt.ylabel('Temperature (ºC)', {'fontsize': 14})
plt.xticks(dft.time[::2])
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10,6))
x_axis = ["1985", "2000", "2020"]
y_axis = [ 26.78, 27.71, 28.73 ]

plt.bar(x_axis, y_axis, color=['black', 'red', 'green'])
plt.ylabel("ºC")
plt.title("SST mean 1985, 2000 and 2021")
plt.show()
```

# Conclution: The maximun Sea Surface Temperature (SST) since 1985 has raised 1.17 ºC in the natural park Islas del Rosario. Since year 2000 the maximum temperature raised 0.72 ºC showing an acceleration over the last 20 years. More dramatic is the SST mean comparision which shows hotter waters with a 1.95 ºC increase from 1985 to now. This area is characterized for a high coral coberture (A.cervicornis) which is a big problem for diseases such as bleaching.
