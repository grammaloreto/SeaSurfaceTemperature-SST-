```python
import pandas as pd
import datetime as dt
```

From Google Earth Engine (GEE) it was extracted the Sea Surface Temperature (SST) data from 11 points of Corales del Rosario National Park/Cartagena/Colombia. The data is basically a daily mean of 14 days from different years. In this case January 1-13 of 1985. With small csv files the aim is to make dataframes to analize.  

```python
dfa = pd.read_csv("cero85.csv")
dfb = pd.read_csv("uno85.csv")
dfc = pd.read_csv("dos85.csv")
dfd = pd.read_csv("tres85.csv")
dfe = pd.read_csv("cuatro85.csv")
dff = pd.read_csv("cinco85.csv")
dfg = pd.read_csv("seis85.csv")
dfh = pd.read_csv("siete85.csv")
dfi = pd.read_csv("ocho85.csv")
dfj = pd.read_csv("nueve85.csv")
dfk = pd.read_csv("diez85.csv")
```
